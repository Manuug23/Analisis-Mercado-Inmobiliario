Conclusiones:

* bathrooms está totalmente lleno con nulos, por tanto se elimina
* description no pasa nada porque tenga nulos, así que se deja
* host_is_superhost tiene muy pocos nulos y no es una variables super relevante, así que se deja
* beds: se puede intentar imputarla a partir de accomodates
* bedrooms sí es una variable importante, se puede intentar imputar los nulos a través de proxies como accomodates o beds, buscando relacion entre la cantidad de personas que entran en un departamento y la cantidad de camas

Se ve si se puede hacer una imputación de beds a partir del número de personas que se pueden acomodar.

```python
pd.crosstab(listings_det.beds, listings_det.accommodates)
```

Parece que sí se podría hacer una asignación mas o menos directa. Leyendo la matriz en vertical se ve que:

* una o dos personas se suelen corresponder con una cama
* tres o cuatro personas se suelen corresponder con dos camas
* cinco o seis personas se suelen corresponder con tres camas
* a más de 6 personas se le vamon a asignar cuatro camas

Se repasa el número de nulos y la frecuencia de cada valor

```python
listings_det.beds.value_counts(dropna = False)
```

```python
Se crea una función para imputar los nulos de beds en base a accommodates
```

```python
def imputar_nulos(registro):
    #lista de condiciones
    condiciones = [(registro.accommodates <= 2), 
                   (registro.accommodates > 2) & (registro.accommodates <= 4),
                   (registro.accommodates > 4) & (registro.accommodates <= 6),
                   (registro.accommodates > 6)]
    #lista de resultados
    resultados = [1,2,3,4]
    #salida
    #select lo que hace es aplicarle a cada condicion de la lista condiciones un valor (en orden) de resultados. Ej si accom es menor = a 2 asigna 1
    return(np.select(condiciones, resultados, default = -999))
#con apply le aplico la funcion imputar_nulos a cada registro de la variable beds que sea NA
listings_det.loc[listings_det.beds.isna(), 'beds'] = listings_det.loc[listings_det.beds.isna()].apply(imputar_nulos, axis = 1).astype('int64')
```

```python
listings_det.beds.value_counts(dropna = False)
```

```python
Ahora se ve si se puede hacer una imputación de bedrooms.

Se comienza por cruzar el número de habitaciones con el número de personas que se pueden acomodar.
```

Se procede a contrastarlo con el número de camas.

Aquí sí se podría hacer una asignación más directa. Leyendo la matriz en vertical se ve que:

* cero, una o dos camas se suele corresponder con una habitación
* tres o cuatro camas se suele corresponder con dos habitaciones
* cinco o seis camas se suele corresponder con tres habitaciones
* a más camas se le van a asignar cuatro habitaciones

Se modifica la función creada para imputar los nulos de bedrooms a partir de beds.

Primero se hace el conteo de bedrooms

```python
listings_det.bedrooms.value_counts(dropna = False)
```

Se ejecuta la función actualizada

```python
def imputar_nulos(registro):
    #lista de condiciones (camas)
    condiciones = [(registro.beds <= 2), 
                   (registro.beds > 2) & (registro.beds <= 4),
                   (registro.beds > 4) & (registro.beds <= 6),
                   (registro.beds > 6)]
    #lista de resultados (habitaciones)
    resultados = [1,2,3,4]
    #salida
    #select lo que hace es aplicarle a cada condicion de la lista condiciones un valor (en orden) de resultados. Ej si beds es menor = a 2 asigna 1
    return(np.select(condiciones, resultados, default = -999))
#con apply le aplico la funcion imputar_nulos a cada registro de la variable bedrooms que sea NA
listings_det.loc[listings_det.bedrooms.isna(), 'bedrooms'] = listings_det.loc[listings_det.bedrooms.isna()].apply(imputar_nulos, axis = 1).astype('float64')
```

```python
listings_det.bedrooms.value_counts(dropna = False)
```

Por último se elimina bathrooms

```python
#elimino la variable bethrooms
listings_det.drop(columns = 'bathrooms', inplace = True)
```

## DATAMART ANALITICO

```python
df = pd.merge(left = listings, right = listings_det, how = 'left', on = 'id')
df.head()
```

Ahora queda agregar al df la variable precio_m2

```python
precio_m2.info()
precio_m2.distrito.unique()
```

```python
df.neighbourhood_group.unique().categories
```

Comparando parece todo igual excepto:

* Fuencarral - El Pardo
* Moncloa - Aravaca
* San Blas - Canillejas

Por tanto se van a reemplazar los valores en precio_m2 para que sean iguales a los de df y se puedan cruzar

```python
precio_m2.distrito = precio_m2.distrito.map({'Fuencarral':'Fuencarral - El Pardo',
                                             'Moncloa':'Moncloa - Aravaca',
                                             'San Blas':'San Blas - Canillejas'}).fillna(precio_m2.distrito)
```

```python
precio_m2.distrito.unique()
```

Ahora si ya se pueden cruzar

```python
df = pd.merge(left = df, right = precio_m2, how = 'left', left_on = 'neighbourhood_group', right_on = 'distrito')
```

```python
df.head()
```

Se comprueba que no se hayan generado nulos en la unión.

```python
df.precio_m2.isna().sum()
```
